# <font color=orange>1) INTRODUCTION</font>
***

The database is made up of administrative migration records collected at border posts and transferred to the Ministry of Tourism by the Federal Police Department. The data are treated statistically according to the theoretical framework of the International Recommendations of Tourism Statistics, edited in 2008, by the World Tourism Organization - WTO, in order to guarantee international comparability. This treatment process begins with the purification of the base, excluding the types of travelers not included in the tourism statistics, followed by the verification of the fit of each record in the tourist classification (visitor who includes an overnight stay in their trip) and conversion the nationality variable in the country of permanent residence, which constitutes a key variable in tourism statistics. In the end, there are the number of international tourists arriving in Brazil.


<font color=orange>SOURCE OF DATA</font>

http://dados.turismo.gov.br/chegada-de-turistas-internacionais

<font color=orange>VARIABLES</font>

- Continent: Continent where the tourist came from
- id continent: Id for the continent
- Country: Country where the tourist came from
- id country: Id for the country
- State: State where the tourist arrived
- id state: id for the state
- Via: via that the tourist arrived
- id via: id for the via
- year: year of arrival
- Month: Month of arrival
- id month: id for the month
- Arrivals: Arrival numbers.
- combination: Column created for the combination about the country of origin and the the state which the tourist arrived.

<font color=orange>OBSERVATION</font>

- Arrival values equal to zero were excluded from the dataset
- The "combination" column was created basically by concatenating the "Country" and "UF" columns
